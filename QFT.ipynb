```python
#QFT
import numpy as np
from numpy import pi
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, transpile, Aer, IBMQ, execute
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

```python
qr = QuantumRegister(3)
cr = ClassicalRegister(3)

qc = QuantumCircuit(qr,cr)

qc.x(2)
qc.h(2)
qc.cp(pi/2, 1, 2) # Rotation gate from qubit 1 to qubit 2
qc.cp(pi/4, 0, 2)
qc.h(1)
qc.cp(pi/2,0,1)
qc.h(0)
qc.swap(0,2)
qc.measure_all()
#qc.measure(qr[1], cr[1])
#qc.measure(qr[2], cr[2])

qc.draw(output='mpl')
```

```python
sv_backend = Aer.get_backend("statevector_simulator") #Choosing a backend
sv_result = execute(qc, sv_backend).result() #submit quantum circuit to backend
#psi = sv_result.get_statevector()
#showing results
counts = sv_result.get_counts()
plot_histogram(counts)
#print(psi)
#qobj = assemble(qc)
#result = simulator.run(qobj).result().get_statevector()
#plot_bloch_multivector(result)
```
